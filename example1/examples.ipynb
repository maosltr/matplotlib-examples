# Plotting data from updating csv file

## 1. Live plot of cpu load (locally)


```python
import shutil
import time
import psutil
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
%matplotlib notebook
plt.rcParams['animation.html']="jshtml"

fig1 = plt.figure()
ax = fig1.add_subplot(111)
fig1.show()
```

```python

while True:  

    shutil.copyfile("demo_csv1.csv", "demo_csv2.csv")

    demo_data = pd.read_csv("demo_csv2.csv")
    ax.plot(demo_data["time"], demo_data["cpu load"], 'bx')
    ax.set(title="demo", 
    xlabel="time", 
    ylabel="cpu load (%)")
    fig1.canvas.draw()
    time.sleep(1)
    i+=0.1
```
